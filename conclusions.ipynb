# Conclusions Final Project

### About the project: 
The most complicated objective and the one I was working on was getting the model to recognize the four classes of camera movement (Static, Motion, Pull, Push). ML models are not designed to be able to understand this type of spatial phenomenon. So I thought of different strategies to implement in the frames that the model can see and better understand what type of class to predict. 

#### Researched and developed:

1. MIDAS library: With this library the model can distinguish the background from the main subject. 

2. Mask with Pytorch and panopticapi.

3. Generate tracking points with OpenCV optical flow.

### 3. POINT TRACKING

Implementing OpenCV's optical flow in a function to pre-process the videos, a dot plot mask was generated. The result was very good and allowed the model to recognize pull and push better than in the scientific paper.

A single training of the model was done for a matter of time (30 epochs) and without using LSTM, so the model only sees a single image per video. This test was done to test the result of the function. 

Not always the function gives good results. If the image is very dark or very light, if the image is very diffuse, or if the video starts with a black or white frame (this was solved by avoiding the first frames of the video) there is no reference to where to follow the traceability and the result is completely lost.
